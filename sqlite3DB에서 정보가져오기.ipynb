```python
import sqlite3

conn = sqlite3.connect('database.db')

cursor = conn.cursor()

sql = """SELECT 상호 FROM basic_info WHERE 업종= '치킨'"""

cursor.execute(sql)

while True:
    row = cursor.fetchone()
    if row == None:
        print('NONE')
        break
    print(row)
    
cursor.close()
conn.close()
```

```python
def getSummaryInfoFromSqlite3DB(name):
    summary_info = []
    
    conn = sqlite3.connect('database.db')

    cursor = conn.cursor()
    
    name = ' '+name

    sql = """SELECT * FROM basic_info WHERE 영업표지= '{}'""".format(name)

    cursor.execute(sql)

    while True:
        row = cursor.fetchone()
        if row == None:
#             print('NONE')
            break
        summary_info = list(row)

    cursor.close()
    conn.close()
    
    return summary_info
```

```python
summary_info = getSummaryInfoFromSqlite3DB('네네치킨')
```
